# GARDIO INTERFACE

```python
import gradio as gr
import numpy as np
import pickle

# Load the trained model
model = pickle.load(open(r'C:\Users\jesti\OneDrive\Desktop\crop_yield.pkl', 'rb'))

# Creating mapping dictionaries
mapping_dict = {
    'Crop': {
        'Arecanut': 0, 'Arhar/Tur': 1, 'Castor seed': 2, 'Coconut ': 3,
        'Cotton(lint)': 4, 'Dry chillies': 5, 'Gram': 6, 'Jute': 7,
        'Linseed': 8, 'Maize': 9, 'Mesta': 10, 'Niger seed': 11,
        'Onion': 12, 'Other  Rabi pulses': 13, 'Potato': 14,
        'Rapeseed &Mustard': 15, 'Rice': 16, 'Sesamum': 17,
        'Small millets': 18, 'Sugarcane': 19, 'Sweet potato': 20,
        'Tapioca': 21, 'Tobacco': 22, 'Turmeric': 23, 'Wheat': 24,
        'Bajra': 25, 'Black pepper': 26, 'Cardamom': 27, 'Coriander': 28,
        'Garlic': 29, 'Ginger': 30, 'Groundnut': 31, 'Horse-gram': 32,
        'Jowar': 33, 'Ragi': 34, 'Cashewnut': 35, 'Banana': 36,
        'Soyabean': 37, 'Barley': 38, 'Khesari': 39, 'Masoor': 40,
        'Moong(Green Gram)': 41, 'Other Kharif pulses': 42, 'Safflower': 43,
        'Sannhamp': 44, 'Sunflower': 45, 'Urad': 46, 'Peas & beans (Pulses)': 47,
        'other oilseeds': 48, 'Other Cereals': 49, 'Cowpea(Lobia)': 50,
        'Oilseeds total': 51, 'Guar seed': 52, 'Other Summer Pulses': 53,
        'Moth': 54
    },
    'Season': {
        'Whole Year ': 0, 'Kharif     ': 1, 'Rabi       ': 2,
        'Autumn     ': 3, 'Summer     ': 4, 'Winter     ': 5
    },
    'State': {
        'Assam': 0, 'Karnataka': 1, 'Kerala': 2, 'Meghalaya': 3,
        'West Bengal': 4, 'Puducherry': 5, 'Goa': 6, 'Andhra Pradesh': 7,
        'Tamil Nadu': 8, 'Odisha': 9, 'Bihar': 10, 'Gujarat': 11,
        'Madhya Pradesh': 12, 'Maharashtra': 13, 'Mizoram': 14, 'Punjab': 15,
        'Uttar Pradesh': 16, 'Haryana': 17, 'Himachal Pradesh': 18,
        'Tripura': 19, 'Nagaland': 20, 'Chhattisgarh': 21, 'Uttarakhand': 22,
        'Jharkhand': 23, 'Delhi': 24, 'Manipur': 25, 'Jammu and Kashmir': 26,
        'Telangana': 27, 'Arunachal Pradesh': 28, 'Sikkim': 29
    }
}

def predict_yield(crop, season, state, area, production, annual_rainfall, fertilizer, pesticide):
    try:
        # Convert categorical strings to numerical values
        crop_num = mapping_dict['Crop'][crop]
        season_num = mapping_dict['Season'][season]
        state_num = mapping_dict['State'][state]

        # Create input array with all numerical values
        input_data = np.array([[
            crop_num, season_num, state_num, area, production,
            annual_rainfall, fertilizer, pesticide
        ]], dtype=np.float32)

        # Predict yield
        predicted_yield = model.predict(input_data)[0]
        return f"Predicted Crop Yield: {predicted_yield:.2f} tons per hectare"
    except Exception as e:
        return f"Error in prediction: {str(e)}"

# Create Gradio interface
inputs = [
    gr.Dropdown(list(mapping_dict['Crop'].keys()), label="Crop", value="Wheat"),
    gr.Dropdown(list(mapping_dict['Season'].keys()), label="Season", value="Rabi       "),
    gr.Dropdown(list(mapping_dict['State'].keys()), label="State", value="Punjab"),
    gr.Number(label="Area (in Hectares)", value=1.0),
    gr.Number(label="Production (in Tons)", value=2.0),
    gr.Number(label="Annual Rainfall (in mm)", value=500.0),
    gr.Number(label="Fertilizer Usage (in kg/ha)", value=100.0),
    gr.Number(label="Pesticide Usage (in kg/ha)", value=10.0)
]

outputs = gr.Textbox(label="Yield Prediction")

gr.Interface(
    fn=predict_yield,
    inputs=inputs,
    outputs=outputs,
    title="Crop Yield Predictor",
    description="Predicting the crop yield based on various factors such as crop type, season, state, rainfall, fertilizer, pesticide"
).launch()
```
